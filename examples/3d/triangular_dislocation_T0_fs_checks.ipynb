# Checks of different solutions for a triangular dislocation

in full-space and half-space


## Simple triangle  // to the plane $e_1,\,e_2$

We compare the outputs of Bigwham for displacement and stresses at a set of observation ploints
with:

1 - direct integration of the full-space fundamental kernels (results from mathematica)

2 - integration of the full-space regularized kernels of Li&Mear (results from mathematica)

3 - Matlab code of Nikkho and Walter 

4 - Matlab code for a TI material of Pan et al  

The results from bigwham are with _bm , the others are copy/paste from the other codes with _1_fs, _2_reg, _3_nw, _4_pan 



```python
import numpy as np
from bigwham4py import BEMatrix

coor = np.asarray([[0.,0.,0.],[2.,0.,0.],[0.,1.,0.]])  # vertex coordinates
conn =np.asarray([[0,1,2]])  # connectivity of the triangular dislocation

G=1.
nu=0.35
kernel = "3DT0-H"
elas_prop = np.array([2*G*(1+nu), nu])

# 4 distincts observations point (inside elemt, above, below and away on plane)
obs_points = np.asarray([[0.3,0.3,0.],[0.3,0.3,1.],[0.3,0.3,-1.],[1.3,2.3,0.]])

h = BEMatrix(kernel, coor, conn, elas_prop, 32, 0., 1e-6,n_openMP_threads=8)
```

```python
# e_1 d
d=np.array([1.,0.,0.])

stress_bm_e1=h.compute_stresses(obs_points,-d)   # minus sign due to convention of positive dd in overlap in the inner of Bigwham....
displ_bm_e1=h.compute_displacements(obs_points,-d)
# e_2 d
d=np.array([0.,1.,0.])

stress_bm_e2=h.compute_stresses(obs_points,-d)   # minus sign due to convention of positive dd in overlap in the inner of Bigwham....
displ_bm_e2=h.compute_displacements(obs_points,-d)
# e_3 d
d=np.array([0.,0.,1.])

stress_bm_e3=h.compute_stresses(obs_points,-d)   # minus sign due to convention of positive dd in overlap in the inner of Bigwham....
displ_bm_e3=h.compute_displacements(obs_points,-d)
```
